```python
import os
import pandas as pd
```

```python
class PlayerCharacter():
    
    def __init__(self,c_name,c_class,c_sub,c_hp,c_lvl,c_primary_weapon,c_armor,ac):
        self.name = c_name
        self.c_class = c_class
        self.sub = c_sub
        self.hp = c_hp
        self.lvl = c_lvl
        self.primary_weapon = c_primary_weapon
        self.armor = c_armor
        self.ac = ac
    
    def character_class(self):
        os.chdir('/Users/MatthewBarnette/DnD_stuff/CSV_files')
        
    
    def weapon(self):
        os.chdir('/Users/MatthewBarnette/DnD_stuff/CSV_files')
        weapons = pd.read_csv('weapons.csv')
        self.primary_weapon_stats = weapons[weapons.Name == self.primary_weapon]
```

```python
class Monster(PlayerCharacter):
    
    def __init__(self,c_name,c_class,c_sub,c_hp,c_lvl,c_primary_weapon,c_armor,ac,cr):
        PlayerCharacter.__init__(self,c_name,c_class,c_sub,c_hp,c_primary_weapon,c_armor,ac,c_lvl)
        self.cr = cr
```

```python
Goblin = Monster(c_name="Goblin_1",c_class="",c_sub="",c_hp=15,cr=.25,c_primary_weapon="Scimitar",c_armor="Leather",ac=15,c_lvl=0)
```
